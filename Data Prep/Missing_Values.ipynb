## KALAU MAKE CONDITIONAL SELECTION, SUMMON DUA KALI (data([data.....]))

```python
# Isi Missing Value horsepower dengan masing-masing mean dari origin
# Mean HP dari USA & Yurop

meanUSA = data[data['origin'] == 'usa']['horsepower'].mean()
meanYurop = data[data['origin'] == 'europe']['horsepower'].mean()
```

```python
data.iloc[32] = data.iloc[32].fillna(meanUSA)
```

```python
data.iloc[32]
```

```python
data[data['horsepower'].isnull()]
```

```python
def HP(x):
    if (x['horsepower'] > 0):
        return(x['horsepower'])
    else:
        if (x['origin'] == 'usa'):
            return meanUSA
        else:
            return meanYurop
```

```python
data['horsepower'] = data[['horsepower','origin']].apply(HP, axis = 1)
```

```python
df1 = sns.load_dataset('titanic')
df1.info()
df1['who']
```

```python
meanMan = int(df1[df1['who'] == 'man']['age'].mean())
meanWoman = int(df1[df1['who'] == 'woman']['age'].mean())
meanChild = int(df1[df1['who'] == 'child']['age'].mean())
```

```python
df1[df1['age'].isnull()]
```

```python
def umur(x):
    if ((x['age'] > 0)):
        return x['age']
    else:
        if (x['who'] == 'man'):
            return meanMan
        elif (x['who'] == 'woman'):
            return meanWoman
        else:
            return meanChild
```

```python
df1['age'] = df1[['age','who']].apply(umur, axis = 1)
```

```python
df1['age']
```

```python
df1 = df1.dropna(subset=['embarked'])
```

```python
df1 = df1.drop(columns = 'deck')
```
