```python
# buat tabel dengan 3 kolom, tahun,jumlah suicide maks, nama negara
tahun = list(df['year'].unique())
negara = []
jumlah = []
for item in tahun:
    z = df[df['year']==item]['suicides/100k pop'].max()
    negara.append(df[(df['year'] == item) & (df['suicides/100k pop'] == z)]['country'])
    jumlah.append(round(df[df['year'] == item]['suicides/100k pop'].max()))
                  
data = {
    'Tahun' : tahun,
    'Negara': negara,
    'Suicide_Max' : jumlah
}
pd.DataFrame(data)
```
