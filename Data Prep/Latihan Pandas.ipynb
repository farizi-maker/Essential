## SERIES

```python
import pandas as pd
import numpy as np
```

```python
list1 = [1,2,3,4,5]
set1 = pd.Series(list1)
```

```python
set1
```

```python
set1[4]
```

```python
set2 = pd.Series(data = list1, index = ['A','B','C','D','E'])
```

```python
set2
```

```python
set2['D']
```

## LIST TO DATAFRAME

```python
ls = [[1,2,3,4],[1,2,3,4]]
pd.DataFrame(data = ls, columns=['A','B','C','D'])
```

## DICTIONARY TO DATAFRAME

```python
d = {
    'nama' : ['Fikri', 'Budi'],
    'alamat' : ['Indonesia', 'Asia']
}
pd.DataFrame(d)
```
