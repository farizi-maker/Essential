```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data= load_breast_cancer()
```

```python
x=pd.DataFrame(data['data'],columns=data['feature_names'])
```

```python
x
```

```python
y=pd.Series(data['target']).map({0:'malignant',1:'benign'})
```

```python
y
```

```python
# noted, must standardscaler
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler_transformer=scaler.fit(x)
```

```python
x_scaled=scaler_transformer.transform(x)
```

```python
x_scaled=pd.DataFrame(x_scaled,columns=x.columns)
```

### MODELING Wo PCA

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
X_train,X_test,y_train,y_test = train_test_split(x_scaled,y,random_state=101)
log_model = LogisticRegression()
log_model.fit(X_train,y_train)
predic=log_model.predict(X_test)
print(classification_report(y_test,predic))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_score = cross_val_score(LogisticRegression(),x_scaled,y,cv=4)
```

```python
cv_score.mean()
```

## Modeling with PCA :)

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
x_PCA_2 = pca.fit_transform(x_scaled)
```

```python
x_PCA_2=pd.DataFrame(x_PCA_2,columns=['PCA 1','PCA 2'])
```

```python
cv_score = cross_val_score(LogisticRegression(),x_PCA_2,y,cv=4)
```

```python
cv_score.mean()
```

## FIND BEST N

```python
cv_scores=[]
N=[]
for i in range(1,30):
    XPCA = PCA(n_components=i).fit_transform(x_scaled)
    cv_score = cross_val_score(LogisticRegression(),XPCA,y,cv=4).mean()
    cv_scores.append(cv_score)
    N.append(i)
    
plt.figure(figsize=(10,4))
plt.plot(N,cv_scores,linestyle='dashed',marker='o')
plt.show()
pd.DataFrame({'N':N,'CV Score':cv_scores}).sort_values(by=['CV Score'],ascending=False)
```

```python
PCA=PCA(n_components=10)
```
