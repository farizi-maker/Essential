```python
import pandas as pd
```

```python
df = pd.read_csv('GPA.csv')
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ln = LinearRegression()
```

```python
ln.fit(df[['SAT']], df['GPA'])
```

```python
ln.intercept_
```

```python
ln.coef_
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1)
ax.scatter(df['SAT'], df['GPA'])
ax.plot(df['SAT'], ln.intercept_ + ln.coef_[0] * df['SAT'], 'r')
```
