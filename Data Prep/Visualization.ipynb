```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(0,5,11)
```

```python
data1 = x
data1
```

```python
y = x**2
data2 = y
data2
```

```python
plt.plot(data1,data2,'r')
plt.xlabel('ini x label')
plt.ylabel('ini y label')
plt.title('ini title dari plot')
```

# Subplot

```python
# parameter nrow,ncol plot number
plt.subplot(1,2,1)
plt.plot(data1,data2)
plt.xlabel('data 1')
plt.ylabel('data 2')

plt.subplot(1,2,2)
plt.plot(data2,data1, 'r')
plt.xlabel('data 1')
plt.ylabel('data 2')
```

# Label,Legend and Title

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(data1,data2,color = 'yellow',label = 'y')
ax.plot(data2,data1,color = 'red', label ='x')
ax.set_title('ini title')
ax.legend()
```

```python
df = sns.load_dataset('tips')
```

```python
df.head(5)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
# axes.scatter(df['total_bill'], df['tip'])
axes.set_xlabel('Total Bill')
axes.set_ylabel('Tip')
axes.set_title('Male vs Female')
axes.scatter(df[df['sex'] == 'Male']['total_bill'], df[df['sex'] == 'Male']['tip'], color = 'red')
axes.scatter(df[df['sex'] == 'Female']['total_bill'], df[df['sex'] == 'Female']['tip'], color = 'blue')
axes.legend(['Male', 'Female'])
```

```python
df['day'].value_counts()
```

```python
day = df['day'].unique()
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
warna = ['red','green','blue','yellow']
for i in range (len(day)):
    axes.scatter(df[df['day'] == day[i]]['total_bill'], df[df['day'] == day[i]]['tip'], color = warna[i],label = day[i])
axes.legend()
```
