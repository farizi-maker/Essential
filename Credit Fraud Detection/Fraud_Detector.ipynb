# Removing Outlier

```python
index_to_drop = df[(df.kode_cabang.isnull()) & (df.flag_kredit_macet == 0)].index
df.drop(index_to_drop, inplace = True)
df['kode_cabang'].fillna(df[df['flag_kredit_macet'] == 1]['kode_cabang'].mode().values[0], inplace = True)
df.flag_kredit_macet.value_counts(normalize = True)
```

```python
def UwU(a) :
    data = a
    listFeature = list(data.columns[1:])
    for item in listFeature:
        q3 = data[item].quantile(0.75)
        q1 = data[item].quantile(0.25)
        iqr = q3-q1
        lower_limit = q1-(1.5*iqr)
        upper_limit = q3+(1.5*iqr)
        return data[(data[item] < lower_limit) | (data[item] > upper_limit) & (data['flag_kredit_macet'] == 0)].index
```

```python
take_out = list(UwU(df))
```

```python
df = df.drop(take_out, axis = 0)
```

```python
df.flag_kredit_macet.value_counts(normalize = True)
```

```python
df['kode_cabang'] = df['kode_cabang'].astype('category')
```

```python
df['kode_cabang'].cat.categories = list(range(0,11))
```

```python
y = df['flag_kredit_macet']
x = df.drop(['flag_kredit_macet', 'X'], axis = 1)
```

# Imbalance Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.metrics import classification_report
```

```python
def awawa (X,Y,model):
    X_train,X_test,y_train,y_test = train_test_split(X,Y)
    clf = model()
    clf.fit(X_train,y_train)
    
    prediction_test = clf.predict(X_test)
    prediction_train = clf.predict(X_train)
    print('Class Report Test Data')
    print(classification_report(y_test, prediction_test))
    print('===========================================================')
    print('Class Report Train Data')
    print(classification_report(y_train, prediction_train))
    print('===========================================================')
```

```python
awawa(x,y, LogisticRegression)
```

```python
awawa(x,y,DecisionTreeClassifier)
```

```python
awawa(x,y,RandomForestClassifier)
```

```python
awawa(x,y,GradientBoostingClassifier)
```

# Handling The Imbalance

# 1. Undersampling

```python
minority_data = df[df['flag_kredit_macet'] == 1]
majority_data = df[df['flag_kredit_macet'] == 0]
```

```python
index_for_mayority = np.random.choice(majority_data.index, len(minority_data))
```

```python
df_class0 = df.loc[index_for_mayority]
```

```python
undersampling_df = pd.concat([df_class0, minority_data])
```

```python
undersampling_df['flag_kredit_macet'].value_counts(normalize = True)
```

```python
under_y = undersampling_df['flag_kredit_macet']
under_x = undersampling_df.drop('flag_kredit_macet', axis = 1)
```

```python
awawa(under_x,under_y,LogisticRegression)
```

```python
awawa(under_x,under_y,DecisionTreeClassifier)
```

```python
awawa(under_x,under_y,RandomForestClassifier)
```

```python
awawa(under_x,under_y,GradientBoostingClassifier)
```
