```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('ISAT.JK.csv', parse_dates = ['Date'])
df2 = pd.read_csv('FREN.JK.csv', parse_dates = ['Date'])
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(df1['Date'], df1['Close'], color = 'yellow', label = 'ISAT')
axes.plot(df2['Date'], df2['Close'], color = 'orange', label = 'FREN')
axes.legend()
axes.set_ylabel('Close Price')
axes.set_xlabel('Date / Daily')
```

```python
plt.subplot(2,1,1)
plt.plot(df1['Date'], df1['Close'], color = 'yellow', label = 'ISAT')
plt.xlabel('Date / Daily')
plt.ylabel('Close')

plt.subplot(2,1,2)
plt.plot(df2['Date'], df2['Close'], color = 'orange', label = 'FREN')
plt.xlabel('Date / Daily')
plt.ylabel('Close')
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (10,8))
axes[0].plot(df1['Date'], df1['Close'], color = 'yellow', label = 'ISAT')
axes[1].plot(df2['Date'], df2['Close'], color = 'orange', label = 'FREN')
axes[0].legend()
axes[1].legend()
```
